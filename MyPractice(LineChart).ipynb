```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weekdays = ['Monday' , 'Tuesday', 'Wednesday' , 'Thrusday', 'Friday']
temperature = [34 , 30.5, 32 ,31.5 ,34 ]
humidity = [31.5 , 33.5 , 32.5 , 31 , 31.5 ]
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title('Weather throughout the weekdays ')
plt.plot(weekdays ,temperature , label = 'Temperature' , linewidth = 1 ,marker = '+', linestyle = '--' , color = 'green')
plt.plot(weekdays ,humidity , label = 'Humidity', linewidth = 2 ,marker = '*', linestyle = 'dotted' , color = 'red')
plt.legend(loc = 'upper center')
plt.show()
```

```python
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title('Weather throughout the weekdays ')
plt.plot(weekdays ,temperature , label = 'Temperature' , marker = 'D' , markersize = 8, linestyle = '' , color = 'green', alpha = 3.5)
plt.plot(weekdays ,humidity , label = 'Humidity', marker = '*'  , markersize = 8, linestyle = '' , color = 'red', alpha = 2.5)
plt.legend(loc = 'upper center')
plt.grid()
plt.show()
```
