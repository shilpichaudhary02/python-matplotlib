```python
import matplotlib.pyplot as plt
```

```python
items = ['Rent', 'phone recharge', 'food' , 'petrol']
spend = [6500 , 800 , 4000 ,1000]
plt.pie(spend, labels =(items) , autopct ='%0.2f%%', explode = [0.2,0,0,0])
plt.show()
```

```python
plt.savefig('Image1.jpeg' , bbox_inches = 'tight')
```
