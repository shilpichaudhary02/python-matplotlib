```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Vertical Bars By Default
weekdays = ['Monday', 'Tuesday', 'Wednesday','Thrusday','Friday']
SoldAccessories = [215 ,150 , 155 , 180 , 120 ]
SoldCars = [120 , 80 , 120 , 80 , 55]
weekdays = np.arange(len(weekdays))
plt.xlabel('Week Days')
plt.ylabel('No. Of SoldItems')
plt.title('Sold Items Throughout The Weekdays')
plt.bar(weekdays+0.2 , SoldAccessories,  width = 0.2, label =  'Sold Accessories' , color = 'grey')
plt.bar(weekdays-0.2 , SoldCars , width = 0.6 , label =  'Sold Cars', color = 'black')
plt.legend(loc = 'best')
plt.show()
```

```python
#Horizontal Bars
plt.xlabel('Week Days')
plt.ylabel('No. Of SoldItems')
plt.title('Sold Items Throughout The Weekdays')
plt.barh(weekdays , SoldAccessories, label =  'Sold Accessories' , color = 'green')
plt.barh(weekdays , SoldCars, label =  'Sold Cars', color = 'black')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```
